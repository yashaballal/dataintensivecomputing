## Import statements

```python
import requests,json
import time
import requests
from bs4 import BeautifulSoup
```

## NYT data extraction code

```python
begindateadd = 0
for j in range(1,100):
    keywords=["anime"]
    for k in keywords:
        print(k)
        begindateadd=(begindateadd+1)%18
        r = requests.get("http://api.nytimes.com/svc/search/v2/articlesearch.json?q="+str(k)+"&begin_date=201901"+str(begindateadd)+"&api-key=YourAPIKeyHere")
        data = r.json()
        file_object = open("NYTDataBeautifulSoup//NYTDataAnimeURL.txt", "ab")
        file_object1 = open("NYTDataBeautifulSoup//NYTDataAnime.txt", "ab")
        if 'fault' in data:
            print("Waiting for 1 minutes for data to be collected again")
            time.sleep(100)
        elif 'response' not in data:
            print("Waiting for 1 minutes for data to be collected again")
            time.sleep(100)            
        else:
            for i in data['response']['docs']:
                    if str(i['web_url']) not in open("NYTDataBeautifulSoup//NYTDataAnimeURL.txt", encoding="utf8").read():
                        print(i['web_url'])
                        file_object.write(i['web_url'].encode("utf8"))
                        page=requests.get(str(i['web_url']))
                        soup = BeautifulSoup(page.text, 'html.parser')
                        anchor_tags=soup.find_all('a')
                        for tag in anchor_tags:
                            para_tag=soup.find_all('p')
                        for tag in para_tag:
                            print("Wrote the data into the file")
                            file_object1.write(tag.prettify().encode("utf8"))
                    else:
                        print("The string was present in news arraylist so not adding")
        file_object.close()
```
