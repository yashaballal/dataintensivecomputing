## Parsing data from MR to retreive the maximum information from the files

```python
dict_max = dict()
counter = 0
for i in range(0,7):    
    file_var = open('TwitterOutputWordCooccurence//part-0000'+str(i), encoding="utf8").read()
    file_var = file_var[file_var.find("MAXIMUM"):]
    for line in file_var.splitlines():
        if line.rfind('|') != -1:
            if '’' not in line and 'relnoopener' not in line and 'evysbk' not in line and 'egzwzxm' not in line and 'eidgb' not in line and 'noreferrer' not in line and 'ejsehar' not in line and 'https' not in line and 'de' not in line and '|t' not in line and '|co' not in line and '|y' not in line and '|s' not in line:
                counter+=1
                key,value = line.rsplit('|',1)
                print(key+" "+value)
                dict_max[key] = int(value)
                if counter > 10:
                    counter-=1
                    del dict_max[min(dict_max, key=lambda k: dict_max[k])]
print(dict_max)
```
