#  PROBLEM 1.1

Creates two lists, one with specific values and the other with values distributed over a Poisson ratio with mean 34. 
It then plots the two variables with choice of colors and value limit. Provides a number of options such as titles, lines, grid, legend to represent the graph data. These options are applied to the latest plot. 

```r
sales1<-c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2<-rpois(12,34)
par(bg="cornsilk")

plot(sales1,col="blue",type="o",ylim=c(0,100),xlab="Month",ylab="Sales")

title(main="Sales by Month")
lines(sales2,type="o",pch=22,lty=2,col="red")
grid(nx=NA,ny=NULL)
legend("topright",inset=.05,c("Sales1","Sales2"),fill=c("blue","red"),horiz=TRUE)
```

# PROBLEM 2

Stores data from the .txt file into the variable sales, and we set the header to true. We display the data read in the form of a barplot, and then as a histogram. Barplot requires input data in the form of matrices or vectors, and since table is not acceptable we need to convert

```r
sales<-read.table("salesdata.txt", header=T)
print(as.matrix(sales))  # to verify that data has been read
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(5))
hist(as.matrix(sales),main="Sales Data",ylab="Total",col="green",border="blue")
```

# PROBLEM 3

In the boxplot function we specify the color to plot each column. We use the stats component of the boxplot to help project the median value which is one of the 5 values of the vector stats.

```r
fn<-boxplot(sales,col=c("orange","green"))$stats
fn
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

# PROBLEM 4

appl1 is the variable which stores the contents of AAPL.csv. Spaces get replaced by dots when read by R,thus the column is specified as Adj.Close, while plotting as well as all the options used for it. The same is done for FB.csv

```r
#fb1<-read.csv("http://real-chart.finance.yahoo.com/table.csv?s=FB&d=10&e=5&f=2014&g=d&a=11&b=12&c=2013&ignore=.csv")
fb1<-read.csv("FB.csv")
aapl1<-read.csv("AAPL.csv")
par(bg="cornsilk")

plot(aapl1$Adj.Close,col="blue",type="o",ylim=c(0,500),xlab="Month",ylab="Sales")
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("red","blue"), horiz=TRUE)
hist(aapl1$Adj.Close, col="red")
hist(fb1$Adj.Close, col="blue")
```

# Problem 6

The google API key that is procured and is used to access the google maps API. Data received from the API i.e. locations of the cities mentioned is then plotted on the map with appropriate symbols. 

```r
install.packages("ggmap")
install.packages("maptools")
```

```r
library("ggmap")
library("maptools")
library("maps")
register_google(key = 'Enter API key') 
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima","Johannesburg,SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

```r
viewport used to tilt the image. Learned from vignette documentation.
```

```r
library(grid)
viewport(x = unit(0.5, "npc"), y = unit(0.5, "npc"),
          width = unit(1, "npc"), height = unit(1, "npc"),
          default.units = "npc", just = "centre",
          gp = gpar(), clip = "inherit",
          xscale = c(0, 1), yscale = c(0, 1),
          angle = 0,
          layout = NULL,
          layout.pos.row = NULL, layout.pos.col = NULL,
          name = NULL)
pushViewport(viewport(h = .8, w = .8, angle = 15))
grid.multipanel(newpage = FALSE)
popViewport()
```
